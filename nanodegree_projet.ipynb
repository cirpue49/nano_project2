```python
#trasfer Sex colum to 0 or 1
def sex_processed(t):
    for i in range(0,len(t)):
        if t["Sex"][i] == "male":
            t["Sex"][i]=0
        if  t["Sex"][i] == "female":
            t["Sex"][i]=1
    return t

sex_processed(data)
```

```python
#Single variable exploration by pie chart
#Focused on Srvived variable I want to predict
%matplotlib inline
import matplotlib.pyplot as plt
survived=0
dead=0
for i in data["Survived"]:
    if i == 1:
        survived+=1
    if i == 0:
        dead+=1
    else:
        continue
fracs=[survived, dead]
labels = 'Survived', 'Dead'
colors= ['gold', 'lightskyblue']
plt.pie(fracs,labels=labels,colors=colors,autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()

#Dead        549   --61.6%
#Survived  342   --38.4%

#Dead value is a bit more than Survived value
```

```python
from scipy import stats 
def correlation(data,variable):
    # Drop na 
    data= data[['Survived', variable]].dropna()   
    return stats.pearsonr(data['Survived'],data[variable])
```

```python
for i in [ 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch',  'Fare']:
    print correlation(data,i)
#From this data exploration, 'Age' and 'Parch' variables are promissing because p-value are less than 0.05( 0.0391 and  0.0147 respectively).
#But, both correlation coefficients are not high. 
```
